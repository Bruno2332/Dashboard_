```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file = "C:/Users/55169/Documents/Python/Projeto_Dados/dados_vendas_ficticios.csv"
```

```python
df = pd.read_csv(file)
```

```python
mais_vendidos = df["Produto"].value_counts()
item_mais_vendido = mais_vendidos.idxmax()
quantidade = mais_vendidos.max()

print(f"O item mais vendido é {item_mais_vendido} com {quantidade} vendas.")
```

```python
plt.figure(figsize=(8,5))
df["Produto"].value_counts().plot(kind="bar", color="skyblue")
plt.title("Itens mais vendidos")
plt.xlabel("Produto")
plt.ylabel("Quantidade de vendas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
